**Test notes:**
- Absorbtion foam acclusing the beam pipe
- Range: 2 MHz to 20 GHz, 10k measurement points
- Power: -5 dBm
- IFBW: 70 kHz
- System calibrated before the measurement 
- Above 8 GHz the system is sensible to the cable movement ! (Probable culprit: the feedthrough; connectors were torque-spannered)

```python
import skrf
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as pat
%matplotlib inline
```

```python
bpm = skrf.Network('./data_VNA_on_spare/TRIUMF.S4P')
```

```python
bpm.plot_s_db()
```

```python
fig, ax = plt.subplots(1, dpi=150)
bpm.s11['2000000.0-10000000000.0 Hz'].plot_s_db(ax=ax, label='S11')
bpm.s22['2000000.0-10000000000.0 Hz'].plot_s_db(ax=ax, label='S22')
bpm.s33['2000000.0-10000000000.0 Hz'].plot_s_db(ax=ax, label='S33')
bpm.s44['2000000.0-10000000000.0 Hz'].plot_s_db(ax=ax, label='S44')
plt.legend()
ax.set_title('VNA analysis')
ax.grid(which='both')

rr = pat.Rectangle((8e9, -40), 2e9,45, color='grey')
ax.add_patch(rr)
plt.savefig('Sxx-reflections.png',format='png',dpi=150)
```

```python
fig, ax = plt.subplots(1, dpi=150)
bpm.s12['2000000.0-10000000000.0 Hz'].plot_s_db(ax=ax, label='S12')
bpm.s13['2000000.0-10000000000.0 Hz'].plot_s_db(ax=ax, label='S13')
bpm.s14['2000000.0-10000000000.0 Hz'].plot_s_db(ax=ax, label='S14')
plt.legend()
ax.set_title('Electrode cross-couplings')
ax.grid(which='both')

rr = pat.Rectangle((8e9, -160), 2e9,165, color='grey')
ax.add_patch(rr)
plt.savefig('Sxx-crosscouplings.png',format='png',dpi=150)
```
